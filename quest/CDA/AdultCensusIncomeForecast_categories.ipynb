### 분석

```python
df_kbi[['gender', 'native_country']].value_counts()
```

```python
df_kbi[['gender', 'native_country']].isna().sum()
```

```python
df_kbi_pivot = pd.pivot_table(data = df_kbi[['gender', 'native_country']]
               , index = 'native_country', columns= ['gender']
               , aggfunc = len).fillna(0) #dropna(), fillna(0) 둘다 pvalue >0.05
df_kbi_pivot
```
